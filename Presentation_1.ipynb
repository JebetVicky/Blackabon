```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor

prs = Presentation()

# Slide 1: Title
slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]
title.text = "Is Contract Farming a Necessary and Sufficient Incentive\nfor Smallholder Sugarcane Farmers in Migori County, Kenya?"
subtitle.text = "Hilver Aoko Odhiambo | University of Nairobi\n\nFocus: Smallholder sugarcane farmers in Migori County\nKey issue: Does contract farming improve profitability?\nApproach: Comparative analysis (contracted vs non-contracted)\nRelevance: Poverty reduction & agricultural development"

notes_slide = slide.notes_slide
notes_slide.notes_text_frame.text = ("Welcome everyone. Today I will present my research proposal which asks whether contract farming is truly a necessary and sufficient incentive for smallholder sugarcane farmers. We focus on Migori County, compare contracted and non-contracted farmers, and link this to poverty reduction and agricultural development.")

# Slide 2: Problem, Objectives, Hypotheses
slide_layout = prs.slide_layouts[1]
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Problem, Objectives & Hypotheses"
content.text = ("Background Reality:\n• Smallholder farmers dominate agriculture but remain among the poorest.\n• Sugarcane is a key livelihood crop in western Kenya.\n• Contract farming provides inputs, market access, and technical support – but farmers report high deductions, delayed payments, and low net returns.\n\nEmpirical Gap: Mixed evidence on profitability effects (positive, negative, or no significant effect).\n\nBroad Objective: Assess effect of contract farming on sugarcane profitability in Rongo Sub-County, Migori.\n\nSpecific Objectives:\n1. Compare gross margins between contracted and non-contracted farmers.\n2. Identify socio-economic factors influencing contract participation.\n\nHypotheses (H₀):\n• No difference in gross margins between groups.\n• Socio-economic factors do not influence contract participation.")
notes_slide = slide.notes_slide
notes_slide.notes_text_frame.text = ("The problem is that despite contract farming being widely promoted, farmers keep complaining about low net returns. Existing studies give conflicting results. My first objective compares profitability directly using gross margin analysis. The second looks at who participates in contracts and why.")

# Slide 3: Literature Insights
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Literature Insights – Contract Farming & Profitability"
content.text = ("What is contract farming?\nA pre-arranged agreement between farmers and buyers specifying production and marketing conditions.\n\nExpected benefits: Better input access, technical support, assured markets, reduced price risk.\n\nKey concerns: Power asymmetries, opaque deductions, delayed payments, rigid delivery requirements, lock-in effects.\n\nEvidence is context-specific:\n• Some studies report income/profit gains under well-designed contracts.\n• Others find no gains or even negative effects when costs and deductions are fully accounted for.\n\nResearch gap: Few farm-level studies on industrial crops like sugarcane comparing contracted vs. non-contracted profitability in specific local contexts (e.g., Migori).")
notes_slide = slide.notes_slide
notes_slide.notes_text_frame.text = ("Literature shows contract farming is not a magic bullet. Benefits depend on contract design, transparency, and enforcement. Most studies focus on high-value crops, not sugarcane. My study fills this gap by focusing on Migori and using detailed cost data.")

# Slide 4: Methodology
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Methodology & Study Area"
content.text = ("Design: Cross-sectional analytical study using primary farm-level data.\n\nTheory: Profit Maximization Theory (MR = MC).\n\nGross Margin (GM) Analysis:\nGM = Total Revenue – Total Variable Cost\n= (Price × Quantity) – (seed + fertilizer + labor + transport costs)\n\nStatistical tools: t-tests (compare group means) + Pearson/Spearman correlations.\n\nStudy area: South Kamagambo Ward, Rongo Sub-County, Migori County.\n\nSample: 60 smallholder sugarcane farmers (contracted and non-contracted) via two-stage sampling.\n\nKey variables: Contract status (1=contracted, 0=non-contracted), Gross margin (KES/ton), Input/labor/transport costs (KES/acre).")
notes_slide = slide.notes_slide
notes_slide.notes_text_frame.text = ("I will use gross margin analysis because it focuses on variable costs and revenue – perfect for short-term farm profitability comparisons. A t-test will tell me if the average gross margin difference between contracted and non-contracted farmers is statistically significant. Correlations will show which cost drivers matter most.")

# Slide 5: Expected Contributions
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Expected Contributions, Work Plan & Significance"
content.text = ("Policy relevance: Evidence for contract design, farmer support programmes, and county/national sugar policies.\n\nDevelopment linkages:\n• SDG 1 (No Poverty) and SDG 8 (Decent Work & Economic Growth)\n• Kenya's Bottom-Up Economic Transformation Agenda (BETA)\n• Migori CIDP priorities (agricultural modernization)\n\nFarmer relevance: Clarifies income effects of joining or avoiding contracts.\n\nBeneficiaries: Smallholder farmers, policymakers, development partners.\n\nWork Plan: 7 months (Oct 2025 – Apr 2026)\nBudget: ~KES 25,000\nKey Outputs: Term paper + BSc degree completion")
notes_slide = slide.notes_slide
notes_slide.notes_text_frame.text = ("This study matters because it directly informs whether contract farming actually improves farmer incomes. If contracted farmers have higher gross margins, then promoting contracts makes sense. If not, we need to rethink contract design or support non-contracted farmers differently. The findings will benefit farmers, policymakers, and development partners.")

prs.save("Hilver_Aoko_Contract_Farming_Presentation.pptx")
print("Presentation saved as 'Hilver_Aoko_Contract_Farming_Presentation.pptx'")
```
